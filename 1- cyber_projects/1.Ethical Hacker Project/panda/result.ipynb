```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
# Replace 'keystroke_log.csv' with the path to your CSV file
df = pd.read_csv("keys.csv", header=None, names=["key"])

plt.style.use("grayscale")

# Process the DataFrame to extract and count occurrences of each key
key_counts = df["key"].value_counts().reset_index()
key_counts.columns = ["key", "count"]

# Sort the data by the frequency in descending order
key_counts = key_counts.sort_values(by="count", ascending=False)

# Plot the data as a line graph
plt.figure(figsize=(10, 6))
plt.plot(
    key_counts["key"], key_counts["count"], color="blue", marker="o", linestyle="-"
)
plt.xlabel("Key")
plt.ylabel("Frequency")
plt.title("Keys Pressed")
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
# Replace 'keystroke_log.csv' with the path to your CSV file
df = pd.read_csv("keys.csv", header=None, names=["key"])
plt.style.use("grayscale")

# Process the DataFrame to extract and count occurrences of each key
key_counts = df["key"].value_counts().reset_index()
key_counts.columns = ["key", "count"]

# Sort the data by the frequency in descending order
key_counts = key_counts.sort_values(by="count", ascending=False)

# Select the top 10 keys
top_10_keys = key_counts.head(10)

# Plot the data as a line graph
plt.figure(figsize=(10, 6))
plt.plot(
    top_10_keys["key"], top_10_keys["count"], color="blue", marker="o", linestyle="-"
)
plt.xlabel("Key")
plt.ylabel("Frequency")
plt.title("Top 10 Most Frequent Keys Pressed")
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.grid(True)

plt.show()
```
